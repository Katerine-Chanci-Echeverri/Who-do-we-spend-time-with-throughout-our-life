# Who do we spend time with across our lifetime?

### As we go through life we build personal relationships with different people ‚Äì family, friends, coworkers, partners. These relationships, which are deeply important to all of us, evolve with time. As we grow older we build new relationships while others transform or fade, and towards the end of life many of us spend a lot of time alone.

# Data Set 

This data set was taken from [Makeover Monday](https://ourworldindata.org/time-use#who-do-we-spend-time-with-across-our-lifetime). 

```python
from google.colab import drive
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

drive.mount("/content/drive")
who = "/content/drive/MyDrive/Colab Notebooks/Coding Dojo üêâ/Semana 4_Explanatory Viz/time-spent-with-relationships-by-age-us.csv"

who_df = pd.read_csv(who)
who_df.head()
```

## Transforming the data for Matplotlib. 

```python
age       = who_df["Year"].values
alone     = who_df.iloc[:, [3]].values
friends   = who_df.iloc[:, [4]].values
children  = who_df.iloc[:, [5]].values
family    = who_df.iloc[:, [6]].values
partner   = who_df.iloc[:, [7]].values
coworkers = who_df.iloc[:, [8]].values
```

## Plot

```python
plt.figure(figsize = (12,8))

alone_plot,     = plt.plot(age, alone, marker = ".", markersize = 8)
friends_plot,   = plt.plot(age, friends, marker = ".", markersize = 8)
children_plot,  = plt.plot(age, children, marker = ".", markersize = 8)
family_plot,    = plt.plot(age, family, marker = ".", markersize = 8)
partner_plot,   = plt.plot(age, partner, marker = ".", markersize = 8)
coworkers_plot, = plt.plot(age, coworkers, marker = ".", markersize = 8)

plt.xlim(left = 15, right = 80)
plt.ylim(bottom = 0, top = 490)
plt.xlabel("Age", fontsize = 18, fontname = "Liberation Sans Narrow")
plt.ylabel("Time (min)", fontsize = 18, fontname = "Liberation Sans Narrow")
plt.title("Who Americans spend their time with, by age", fontsize = 20, 
          fontname = "Liberation Sans Narrow", fontweight = "bold")
plt.xticks(fontsize = 16);
plt.yticks(fontsize = 16);
plt.grid(axis = "y", c = "white", linestyle = "--")
plt.grid(axis = "x",  alpha = 0, linestyle = "--")
plt.legend([alone_plot, friends_plot, children_plot, family_plot, partner_plot, 
            coworkers_plot], ["Alone", "With friends", "With children", 
                              "With family", "With partner", "With coworkers"],
           prop = {'size': 15});

plt.show();
```
